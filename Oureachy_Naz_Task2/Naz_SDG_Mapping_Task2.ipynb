### WORLD BIODIVERSITY HOTSPOT

```python
world_biodiversity = gpd.read_file(r'/home/naza/Downloads/CI_BiodiversityHotspots.geojson')
world_biodiversity.head()
```

```python
world_biodiversity = gpd.read_file(r'/home/naza/Downloads/CI_BiodiversityHotspots.geojson')
world_biodiversity
```

```python
# Functions defined for mapping each datasets
# For Biodiversity Hotspot
def map_biodiversity_hotspot_to_sdg(row):
    # Some biodiversity hotspots using the "NAME" row from the list above
    marine_biodiversity_hotspots = [
        'Chilean Winter Rainfall and Valdivian Forests', 
        'Indo-Burma', 
        'Japan', 
        'Mesoamerica', 
        'Horn of Africa',
        'Sundaland',
        'Mediterranean Basin',
        'Coastal Forests of Eastern Africa'
    ]
    
    # verify whether this hotspot is known for marine or land biodiversity
    if row['NAME'] in marine_biodiversity_hotspots:
        return "SDG #14: Life below Water"
    else:
        return "SDG #15: Life on Land"
```

```python
# Implement the mapping function defined and save the datasets 
world_biodiversity['SDG_MAPPING'] = world_biodiversity.apply(map_biodiversity_hotspot_to_sdg, axis=1)
world_biodiversity.to_file('CI_BiodiversityHotspots_Mapped.geojson', driver='GeoJSON')
#world_biodiversity
world_biodiversity.head()
```

```python
# Data visualisation using Bar Chart
sdg_biodiversity_count = world_biodiversity['SDG_MAPPING'].value_counts()
plt.figure(figsize=(12, 6))
sdg_biodiversity_count.plot(kind='bar', color=['brown', 'peru'])
plt.title('Distribution of SDGs in Biodiversity Hotspots')
plt.xlabel('Sustainable Development Goals')
plt.ylabel('Number of Biodiversity Hotspots')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

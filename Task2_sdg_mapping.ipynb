# Mapping Nigeria Soil Resources to specific Sustainable Development Goals (SDGs)

```python
# Mapping soil types in Nigeria to sdgs
def map_NGA_soil_to_sdg(row):
    if row ['IPCC'] == 'Acrisol':
        return "SDG 2: Zero Hunger"
    elif row['IPCC'] != 'Waterbodies':
        return "SDG 15: Life on Land"
    else:
        return "Not Mapped"
```

```python
climate = gpd.read_file(r'C:\Users\user\Documents\OUTREACHY\climate\IPCC_ClimateZoneMap_Vector.geojson')
climate
```

```python
#print the climate in Nigeria
NGA_dataset = gpd.read_file (r'C:\Users\user\Documents\OUTREACHY\NGA_AL2_Nigeria.json')
nigeria_climate = gpd.clip(climate, NGA_dataset)

nigeria_climate
```

```python
# Print the data in the 'CLASS_NAME' column
class_names = nigeria_climate['CLASS_NAME']
print(class_names)
```

From the climate metadata, these are the information provided for the CLASS_NAME column

1. Warm temperate moist
2. Warm temperate dry
3. Cool temperate moist
4. Cool temperate dry
5. Polar moist
6. Polar dry
7. Boreal moist
8. Boreal dry
9. Tropical montane

```python
# Count the occurrences of each value in the 'CLASS_NAME' column
value_counts = nigeria_climate['CLASS_NAME'].value_counts()

# Display the counts
print(value_counts)
```

```python
# this code aim to map two relevant SDGs to the row['CLASS_NAME']
def map_nigeria_climate_to_sdg(row):
    if row['CLASS_NAME'] == 9:
        return "SDG 6: Clean Water and Sanitation"
    else:
        return "Not Mapped"
```

```python
NGA_GEZ = gpd.read_file(r'C:\Users\user\Documents\OUTREACHY\NGA_AL2_Nigeria_GEZ.json')
NGA_GEZ
```

```python
gez_name = NGA_GEZ['gez_name']
gez_name
```

# Mapping geoecological zones to specific Sustainable Development Goals (SDGs)

```python
# this code aim to map two relevant SDGs to the row['gez_name']
def map_NGA_GEZ_to_sdg(row):
    if row['gez_name'] != 'Water':
        return "SDG 15: life on Land"
    else:
        return "SDG 14: Life below Water"
```

```python
#Create and apply a mapping column in the dataset
NGA_soil['SDG_Mapping'] = NGA_soil.apply(map_NGA_soil_to_sdg, axis = 1)
nigeria_climate['SDG_Mapping'] = nigeria_climate.apply(map_nigeria_climate_to_sdg, axis = 1)
NGA_GEZ['SDG_Mapping'] = NGA_GEZ.apply(map_NGA_GEZ_to_sdg, axis = 1)
```

```python
# Save the mapped dataset
NGA_soil.to_file("Mapped_Nigeria_Soil_Resources.geojson", driver="GeoJSON")
nigeria_climate.to_file("Mapped_Nigeria_IPCC_ClimateZoneMap.geojson", driver="GeoJSON")
NGA_GEZ.to_file("Mapped_Nigeria_GEZ.geojson", driver="GeoJSON")
```

# Visualize the Mapping

```python
#create a bar chart for Nieria Soil Resources
sdg_soil_counts = NGA_soil['SDG_Mapping'].value_counts()

plt.figure(figsize=(8, 6))
sdg_soil_counts.plot(kind='bar', color=['green', 'blue', 'red']) 
plt.title('Mapping of Nigeria Soil Resources to SDGs')  
plt.xlabel('SDGs')
plt.ylabel('Number of Nigeria Climate Zones') 
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show()  
```

```python
#create a bar chart for Nieria CLimate Zones
sdg_climate_counts = nigeria_climate['SDG_Mapping'].value_counts()

plt.figure(figsize=(8,6))
sdg_climate_counts.plot(kind='bar', color=['skyblue', 'red']) 
plt.title('Mapping of Nigeria Climate Zones to SDGs')  
plt.xlabel('SDGs')  # Label for the x-axis
plt.ylabel('Number of Nigeria Climate Zones') 
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show() 
```

```python
#create a bar chart for Nigeria GeoEcological Zones
sdg_GEZ_counts = NGA_GEZ['SDG_Mapping'].value_counts()

plt.figure(figsize=(8,6))
sdg_GEZ_counts.plot(kind='bar', color=['green', 'skyblue']) 
plt.title('Mapping of Nigeria GeoEcological Zones to SDGs')  
plt.xlabel('SDGs')  # Label for the x-axis
plt.ylabel('Number of Nigeria GEZ') 
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()  # Improve layout
plt.show() 
```

## Title: GFSAD 30m Cropland Extent

### Description
Global Food Security-support Analysis Data (GFSAD) Cropland Extent data from the GFSAD30 product.<br>
The NASA Making Earth System Data Records for Use in Research Environments (MEaSUREs) Global Food Security-support Analysis Data (GFSAD) data product provides cropland extent data for nominal year 2015 at 30 meter resolution (GFSAD30). The monitoring of global cropland extent is critical for policymaking and provides important baseline data that are used in many agricultural cropland studies pertaining to water sustainability and food security. The GFSAD30 data product uses the pixel-based supervised classifier, Random Forest (RF), to retrieve cropland extent from a combination of Landsat 8 Operational Land Imager (OLI) and elevation derived from the Shuttle Radar Topography Mission (SRTM) Version 3 data products. Each GFSAD30 GeoTIFF file contains a cropland extent layer that defines areas of cropland, non-cropland, and water bodies over a 10째 by 10째 area across 7 regions - NACE North America, SACE South America, AUNZCHMOCE Australia, New Zealand, China, and Mongolia, AFCE Africa, EUCEARUMECE Europe, SAAFGIRCE South Asia, Afghanistan and Iran and SEACE South East Asia<br>

Note the FLINT ready data has been resampled from 30m to 0.00025degrees.

### Processing
Resampling to 0.00025 resolution completed in batch as per code below. Use this code to resample tiles or mosaics to another resolution.

As for preprocessing from the original, as the original tiled data is not padded to the nearest degree - best practice is to mosaic (or virtual raster) at original resolution (in sections if computing power is a problem) then gdalwarp or export to tif rounding the bounding extents to the nearest 10th degree, and nearest neighbour resample to 0.00025 or other resolution. From here - these can be tiles using gdal retile making each tile 40000x40000 pixel height and width.

```python
# Batch resample resolution all tiles gdal (this will not force the tiles to sit on the nearest degree)
for /R C:\inputfolder\GFSAD\ %f IN (*.tif) do gdalwarp.exe -co COMPRESS=DEFLATE -co PREDICTOR=2 -co ZLEVEL=9 -s_srs EPSG:4326 -tap -tr 0.00025 0.00025 -multi -r near -of GTiff %f C:\outputfolder\gfsad\tiles\%~nf.tif
```

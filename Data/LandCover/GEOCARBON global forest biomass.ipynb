## Title: GEOCARBON global forest biomass map

### Description
Global map of aboveground biomass in forest, generated by combining and harmonizing the Avitabile pan-tropical biomass map with the boreal forest biomass map by Santoro et al. (2015).<br>

A first global forest biomass map: The pan-tropical biomass map was integrated with the boreal forest biomass component (Santoro et al., 2015) into the GEOCARBON global aboveground forest biomass map at 0.01° resolution. This is the first global biomass map based on field observations and remote sensing data. The map covers only forest areas, where forest are defined as areas with dominance of tree cover in the GLC2000 map (Bartholomé and Belward, 2005).

### Processing
Dataset is already in FLINT appropriate format. To resample/change the pixel resolution from 0.01 degree, please use the following code.

```python
#This dataset has been realigned with the full global extent 180, 90, -90, -180 for use in FLINT but the resolution remains native. Use this code to resample a different raster resolution at -tr using GDAL (or QGIS)
gdalwarp.exe -co COMPRESS=DEFLATE -co PREDICTOR=2 -co ZLEVEL=9 -t_srs EPSG:4326 -tr 0.01 0.01 -r near -te -180.0 -90.0 180.0 90.0 -te_srs EPSG:4326 -of GTiff input.tif output.tif
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import glob
```

```python
# Define the URL of the web page with the iframe
url = "https://unstats.un.org/sdgs/dataportal/countryprofiles/nga?goal=15&series=AG_LND_FRST,ER_PTD_FRHWTR,ER_PTD_TERR,AG_LND_FRSTBIOPHA,AG_LND_FRSTCERT,AG_LND_FRSTCHG,AG_LND_FRSTMGT,AG_LND_FRSTPRCT,AG_LND_DGRD,ER_PTD_MTN,ER_MTN_GRNCVI,ER_RSK_LST,ER_CBD_SMTA"
```

```python
# Send an HTTP GET request to the URL
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the specific element(s) related to SDG 15
    sdg15_element = soup.select('div.sdg-data-sdg15')

    # Extract the text or data from the element
    sdg15_data = sdg15_element[0].csv if sdg15_element else "Data not found"

    print("SDG 15 Data:", sdg15_data)
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

```python
# Create an empty list to store DataFrames
dfs = []

# Specify the folder where your CSV files are located
folder_path =  "file:///home/SwizzCoder/Outreachy_SwizzCoder_2023/SDG15"

# Use glob to get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Loop through each CSV file and load it into a DataFrame
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
df_land = pd.concat(dfs, ignore_index=True)

# Now df_land contains data from all CSV files in the folder
```

```python
# Use glob to get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Loop through each CSV file and process it
for csv_file in csv_files:
    # Extract the base name of the CSV file (without the path)
    file_name = os.path.basename(csv_file)

    # Read the CSV file into a DataFrame
    df_land = pd.read_csv(csv_file)

    # Perform operations on the DataFrame for the current CSV file
    # For example, you can print the first few rows:
    print(f"Data from {file_name}:")
    print(df_land.head())

    # Create a line graph for SDG 15 with the file name as the title
    plt.plot(df_land["TimePeriod"], df_land["Value"], label=file_name)
    plt.xlabel("TimePeriod")
    plt.ylabel("Value")
    plt.title(f"SDG 15: Life on Land - {file_name}")
    plt.xticks(rotation=45)
    plt.show()

    # A separator for clarity between files
    print("-" * 50)
```

The Above Prototype makes use of SDG-15 as it has data on Nigeria related to the land sector repository... the SDG site didn't have data for SDG-13 and so the analysis couldn't be carried out... the link to the SDGs for Nigeria is "https://unstats.un.org/sdgs/dataportal/countryprofiles/nga#goal-15"

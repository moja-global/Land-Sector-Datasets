# SPATIAL QUERY

Spatial query is just to select hotspots that intersect with specific geographic features or regions. In this task, we will check the hotspots that intersects with Nigeria boundary

```python
#spatial Queries
#this code is to detect hotspots that intersect with Nigeria's boundary

hotspots_in_nigeria = gpd.sjoin(biodiversity, NGA_dataset, predicate='intersects', how='inner')
hotspots_in_nigeria
# 'hotspots_in_nigeria' now contains the hotspots that intersect with Nigeria's boundaries

```

# DENSITY ANALYSIS

Density analysis aim to calculate the density of the biodiversity hotspots. It is useful for understanding where hotspots are concentrated

```python
# Calculate the total area covered by all biodiversity hotspots (in square kilometers)

# filter out the hotspot areas

total_area_covered  = biodiversity['Shape_Area'].sum()/1e6 # Convert to square kilometers

# Calculate the total number of biodiversity hotspots
total_hotspot_count = len(biodiversity)

# Calculate the density of biodiversity hotspots over the entire dataset
total_hotspot_density = total_hotspot_count / total_area_covered

print(f"Total Biodiversity Hotspot Density: {total_hotspot_density:.2f} hotspots per square kilometer")
```

```python
# Plot the biodiversity dataset
ax = biodiversity.plot(edgecolor='black', facecolor='none', linewidth=1)

plt.title(f"Biodiversity Hotspot Density\n({total_hotspot_density:.2f} hotspots per square kilometer)")

# Show the plot
plt.show()
```

# Spatial AutoCorrelation

```python
from esda.moran import Moran
```

```python
# Reproject the dataset to World Mercator
biodiversity = biodiversity.to_crs(epsg=3395)
attribute_to_analyze = 'Shape_Area'

data_series = biodiversity[attribute_to_analyze]

#Create a spatial weights matrix (W) to represent the spatial relationships between features. You can use the Queen contiguity:
from libpysal.weights import Queen

w = Queen.from_dataframe(biodiversity)
```

The Moran's I statistic quantifies spatial autocorrelation, and the Moran scatterplot helps you visualize the patterns in the dataset

```python
#Calculate Moran's I to assess spatial autocorrelation:

moran = Moran(data_series, w)

# Display Moran's I statistic
print(f"Moran's I: {moran.I}")
print(f"P-Value: {moran.p_sim}")
```

```python
from splot.esda import moran_scatterplot

# Create a Moran scatterplot
fig, ax = moran_scatterplot(moran, p=0.05)
plt.title(f'Moran Scatterplot for {attribute_to_analyze}')
plt.show()
```

# Spatial Overlay of World Soil Resources on Biodiversity Hotspots

```python
# Perform a spatial overlay to identify intersections

biodiversity = biodiversity.to_crs(epsg=3395)

intersections = gpd.overlay(biodiversity, world_soil, how='intersection')

# Calculate the area of the intersections
intersections['intersection_area'] = intersections['geometry'].area

# Group the data by soil type and calculate the total intersection area for each soil type
soil_type_areas = intersections.groupby('IPCC')['intersection_area'].sum()

# Print the result
print(soil_type_areas)
```

```python

# Plot the world_soil dataset
world_soil.plot(color='lightgray', figsize=(12, 12), alpha=0.7, label="World Soil")

# Plot the biodiversity dataset
biodiversity.plot(color='green', alpha=0.5, label="Biodiversity")

# Plot the intersections
intersections.plot(color='blue', alpha=0.6, label="Intersections")


# Add titles and labels
plt.title('Intersection of World Soil and Biodiversity Datasets')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```

```python
# Perform a spatial overlay to identify intersections
intersections = gpd.overlay(biodiversity, world_soil, how='intersection')

# Calculate the area of the intersections and other statistics
intersections['intersection_area'] = intersections['geometry'].area

# Calculate summary statistics
statistics = intersections.groupby('IPCC')['intersection_area'].describe()

# Print the summary statistics
print(statistics)
```

```python
# Calculate summary statistics
soil_stats = intersections.groupby("IPCC")["intersection_area"].describe()

# Extract statistics data
statistics = soil_stats.loc[:, ["mean", "std", "min", "25%", "50%", "75%", "max"]]

# Transpose the DataFrame for better visualization
statistics = statistics.T

# Create a bar plot
ax = statistics.plot(kind='bar', figsize=(12, 6))

# Customize the plot
ax.set_title("Summary Statistics for Intersection Areas by Soil Type")
ax.set_ylabel("Intersection Area")
ax.set_xlabel("Statistic")
ax.legend(title="Soil Type", loc="upper left", bbox_to_anchor=(1, 1))

plt.show()
```

The distribution and variation in these statistics provide insights into the diversity and spread of different soil types in the regions where they intersect with biodiversity data. Some soil types are more widespread, while others are more concentrated in specific areas. The standard deviation also tells us how much the intersection areas for each soil type deviate from their respective means. This information is valuable for understanding the spatial characteristics of soil types in relation to biodiversity.ach soil type.
